## Task #6
Business wants to check the Claim Amount for deviation for each transaction, they would
like you to calculate as follow
CLAIM_DEVIATION = AVG_CLAIM_AMOUNT_FOR_LAST_30DAYS (same insurance type)
/ CURRENT_CLAIM_AMOUNT
If the value < 0.5 THEN CLAIM_DEVIATION = 1 ELSE 0
If there is less than 30 days of transaction history THEN
-1 Note: LAST_30DAYS does not include current day

```python
merged_ins_emp_ven_df['TXN_DATE_TIME'].max()
```

```python
curr_date = pd.to_datetime('2021-06-30').date()
from datetime import timedelta
# date before 30 days
date_before_30_days = str(curr_date - timedelta(30))
curr_date = str(curr_date)
```

```python
merged_ins_emp_ven_df['TXN_DATE_TIME'] = pd.to_datetime(merged_ins_emp_ven_df['TXN_DATE_TIME'])

merged_ins_emp_ven_df['TXN_DATE_TIME']
```

```python
# df with last 30 days data
last_30days_data_df = merged_ins_emp_ven_df.loc[(merged_ins_emp_ven_df['TXN_DATE_TIME'] >= date_before_30_days ) & (merged_ins_emp_ven_df['TXN_DATE_TIME'] < curr_date)]  
last_30days_data_df
```

```python
average_claim = last_30days_data_df.groupby('INSURANCE_TYPE').count()
average_claim
```

```python
# Getting average claim of last 30 days insurance type wise and storing it in dictionary
average_claim = last_30days_data_df.groupby('INSURANCE_TYPE').mean()['CLAIM_AMOUNT'].to_dict()
average_claim
```

```python
# defining a function for calculating claim_deviation
def Claim_deviation(X):
    if (pd.to_datetime(curr_date) - X['TXN_DATE_TIME'])< timedelta(days=30):

        return -1
    claim_deviation = average_claim[X['INSURANCE_TYPE']] / X['CLAIM_AMOUNT']
    if claim_deviation < 0.5:
        return 1
    else:
        return 0
```

```python
# Applying the function to the dataframe
merged_ins_emp_ven_df['CLAIM_DEVIATION'] = merged_ins_emp_ven_df.apply(Claim_deviation,axis=1)
merged_ins_emp_ven_df['CLAIM_DEVIATION']
```

```python
merged_ins_emp_ven_df.CLAIM_DEVIATION.value_counts()
```

## Task #8
Mobile & Travel Insurance premium are discounted by 10%

Health and Property Insurance premium are increased by 7%

Life and Motor Insurance premium are marginally increased by 2%

What will be overall change in % of the Premium Amount Collected for all these Customer?

```python
Premium_before_dis_inc = merged_ins_emp_ven_df['PREMIUM_AMOUNT'].sum()
Premium_before_dis_inc
```

```python
# total premium for each insurance type
prem_by_types = merged_ins_emp_ven_df.groupby('INSURANCE_TYPE').sum()['PREMIUM_AMOUNT']
prem_by_types
```

```python
# 10% discounted on Mobile & Travel premium
prem_by_types['Mobile'] = np.multiply(prem_by_types['Mobile'],0.9)
prem_by_types['Travel'] = np.multiply(prem_by_types['Travel'],0.9)
# 7% increases on Health and Property premium 
prem_by_types['Health'] = np.multiply(prem_by_types['Health'],1.07)
prem_by_types['Property'] = np.multiply(prem_by_types['Property'],1.07)
# 2% increased on Life and Motor premium increased
prem_by_types['Life'] = np.multiply(prem_by_types['Life'],1.02)
prem_by_types['Motor'] = np.multiply(prem_by_types['Motor'],1.02)
```

```python
#premimum after dicount and increae
premium_after__dis_inc = sum(prem_by_types.to_dict().values())
premium_after__dis_inc
```

```python

change_in_premium = premium_after__dis_inc-Premium_before_dis_inc
change_in_premium
```

```python
# percentage change in premium after discount and increase
percentage_change = ((premium_after__dis_inc-Premium_before_dis_inc)*100)/(Premium_before_dis_inc)
np.round(percentage_change,2)
```
